```python
from sympy import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
beta, I, sigmaI, I0 = symbols(r'beta I sigma_I I_0', real=True)
```

```python
p_bw = beta * I ** (beta-1)
p_bw
```

```python
p_sc = 1/(I*sigmaI*sqrt(2*pi)) * exp(-(log(I/I0) + sigmaI**2 / 2)**2 / (2*sigmaI**2))
p_sc
```

```python
pcom = integrate(p_sc, (I0, 0, 1)).doit()
pcom18 = pcom.subs(sigmaI, 18).evalf() * p_bw
# pcom18 = pcom.subs(sigmaI, 18)
pcom18
```

```python
class Pcom(st.rv_continuous):
    def _pdf(self, x, bet):
        return lambdify([I, beta], pcom18)(x, bet)

    # def _argcheck(self, bet, sigma_I):
    #
    #     s = sigma > 0
    #     l = -1 < lam < 1
    #     p_bool = p > 0
    #     q_bool = q > 0
    #
    #     all_bool = s & l & p_bool & q_bool
    #
    #     return all_bool
```
